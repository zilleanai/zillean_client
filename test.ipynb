```python
## classifier api
from mlplatform_client.api.classifier import classifier as ClassifierApi
project = Project(apihost='mnist.chriamue.de', name='mnist', api='/api/v1')
classifier_api = ClassifierApi(project)
```

```python
from mlplatform_client.api.image import image as ImageApi
image_api = ImageApi(project)
image_files = image_api.list()
img_file = image_files[1]
```

```python
img = image_api.download(img_file)
img
```

```python
from mlplatform_client.api.file import file as FileApi
file_api = FileApi(project)
tags = file_api.tags(img_file)
tags
```

```python
prediction = classifier_api.predict(img)
prediction
```
